12. What is correlation?

  - Correlation is a statistical measure that shows the relationship between two variables. It tells us whether two variables move in the same direction, opposite directions, or have no relationship at all. The correlation value ranges from -1 to +1.

  +1: Perfect positive correlation
  0: No correlation
  -1: Perfect negative correlation

15. What is causation? Explain difference between correlation and causation with an example.

  - Causation means that one event directly causes another event to happen. It represents a true cause-and-effect relationship.

  Difference:
   > Correlation means two variables are related.
   > Causation means one variable directly affects the other.

  Example: Ice cream sales and drowning cases both increase during summer. They are correlated, but eating ice cream does not cause drowning. The actual cause is the hot weather, which increases both ice cream consumption and swimming activities.

16. What is an Optimizer? What are different types of optimizers? Explain each with an example.

  - An optimizer is an algorithm that updates the model's parameters to reduce the loss and improve prediction accuracy. It helps the model learn efficiently during training.

   Some common optimizers are:

   > Gradient Descent: Updates parameters using the entire dataset.
   
   > Stochastic Gradient Descent (SGD): Updates parameters using one training example at a time, making it faster.
   
   > Mini-Batch Gradient Descent: Uses a small batch of data for each update, balancing speed and accuracy.
   
   > Adam Optimizer: Combines the advantages of SGD and adaptive learning rates. It is one of the most widely used optimizers in deep learning because it converges quickly.

   Example: Adam optimizer is commonly used while training neural networks because it provides faster and more stable learning.

21. What is feature scaling? How does it help in Machine Learning?

  - Feature scaling is the process of bringing numerical features to a similar range or scale. Different features may have different units and values, which can affect the performance of some machine learning algorithms.

  - Feature scaling helps by:

   > Improving model accuracy.
   > Speeding up model training.
   > Preventing features with larger values from dominating others.
   > Helping optimization algorithms converge faster.

  - Common methods include Standardization and Normalization (Min-Max Scaling).

25. Explain data encoding.

  - Data encoding is the process of converting categorical (text) data into numerical values so that machine learning algorithms can understand and process it. Since most ML models work only with numbers, encoding is an important preprocessing step.

  - Some common encoding techniques are:

  - Label Encoding: Assigns a unique number to each category.
  - One-Hot Encoding: Creates separate binary columns for each category.
  - Ordinal Encoding: Used when categories have a meaningful order.
  - Target Encoding: Replaces categories with the average value of the target variable.
